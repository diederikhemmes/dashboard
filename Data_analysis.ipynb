```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.titlesize'] = 15
```

```python
url_bank1 = 'https://docs.google.com/spreadsheets/d/17t8Xa8IYHKu2ITComh7CrbDFZm3mvm1pk4VcOnema54/export?&format=csv'
url_bank2 = 'https://docs.google.com/spreadsheets/d/17yS0JJTaQbmwrsIgm71phwoEhQw6hiGfAb5djccP_Fo/export?&format=csv'

df_bank1 = pd.read_csv(url_bank1, decimal=',')
df_bank2 = pd.read_csv(url_bank2, decimal=',')
df_total = pd.concat([df_bank1, df_bank2], ignore_index=True)
display(df_total.head())
```

```python
mean_score = df_total['Final score'].mean()
fig, ax = plt.subplots(figsize=(8,6))
ax.hist(df_total['Final score'])
ax.set_xlabel('Circularity Risk Score')
ax.set_ylabel('Count')
ax.set_title('Overview of Circularity Risk Scores')
print(mean_score)
```
