# Data analysis of Circular Risk Scorecard input
This notebook analyses the data obtained by the Circular Risk Scorecard. 

```python
'''
Define imports and global plot parameters.
'''

import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np

plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.titlesize'] = 15
plt.style.use('tableau-colorblind10')   
```

## Read in data from Sheet

```python
'''
Please fill this in. 
'''

# Provide path to CSV
path_to_csv = r"C:\Users\IrisReitsmaRiskQuest\Documents\GitHub\ScoreCardData_1 - Blad1.csv"

# If True, open only test data. If False, open real data. 
test_data = True 

# Version results
version = 'v0.61' 
```

```python
# Open CSV
df = pd.read_csv(path_to_csv, decimal=',')
display(df.head())

# Select target rows
df_total = df[(df['Version']==version) & (df['Test']==test_data)]
```

## Overview of Circular Risk scores

```python
# Plot distribution of circular risk scores
fig, ax = plt.subplots(figsize=(8,6))
ax.hist(df_total['Final score'], bins=25)
ax.set_xlabel('Circularity Risk Score')
ax.set_ylabel('Count')
ax.set_title('Overview of Circular Risk scores')

# Print mean circular risk score
mean_score = round(df_total['Final score'].mean(), 2)
print('Mean of circular risk scores: ' + str(mean_score))
```

## Relation between Circular Risk Score and Internal PDs

```python
# Scatter plot circular risk scores and internal PDs
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(df_total['Final score'], df_total['Internal score'])
ax.set_xlabel('Circularity Risk Score')
ax.set_ylabel('Internal PD Score')
ax.set_title('Relation between Circular Risk Score and Internal PDs')
plt.show()
```

## Overview of overwrites

```python
# Loop over risk drivers
for i in range(6):
    var_name = 'overwrite ' + str(i + 1)  

    # Count overwrites
    counts = len(df_total[df_total[var_name]==True])

    # Count percentage of overwrites
    total_rows = len(df_total)
    percentage = round(counts/total_rows * 100, 1)

    # Print information
    print('Overwrites for risk driver ' + str(i + 1) + ': ' + str(counts) + ' (=' + str(percentage) + '%)')
```

## Overview of users

```python
# Obtain timestamps and banks
timestamps = df_total['Date'] + ' ' + df_total['Time']
categories = df_total['Bank']

# Convert timestamps to datetime objects
date_objects = [datetime.strptime(ts, "%Y-%m-%d %H:%M:%S").date() for ts in timestamps]

# Create a DataFrame from the data
data = {'Date': date_objects, 'Banks': categories}
df_users = pd.DataFrame(data)

# Count the number of submits per day and category
daily_category_counts = df_users.groupby(['Date', 'Banks']).size().unstack(fill_value=0)

# Plotting
plt.figure(figsize=(10, 6))
daily_category_counts.plot(kind='bar', stacked=True)
plt.title('Daily Submits by Bank')
plt.xlabel('Date')
plt.ylabel('Submit Count')
plt.xticks(rotation=45)
plt.legend(title='Bank')
plt.grid(axis='y')
plt.tight_layout()

# Show the plot
plt.show()

# Display total submit count
print('Total submits for each bank:')
display(daily_category_counts.sum())
```

## Score distributions per risk driver

```python
def plot_histograms_in_row(data_lists, titles=None, bins=10):
    """
    Plot histograms based on input lists in a row of figures.
    
    Parameters:
    data_lists (list of lists): Input data for histograms.
    titles (list of str): Titles for individual histograms. If None, no titles are shown.
    bins (int): Number of bins for histograms.

    Returns:
    None
    """

    # Determine number of subplots
    num_plots = len(data_lists)
    fig, axes = plt.subplots(1, num_plots, figsize=(5 * num_plots, 5))

    # Loop over variables
    for i, data in enumerate(data_lists):

        # Adjust color
        color='C0'
        if titles[i][0:3] == 'tot':
            color = 'C1'

        # Plot histogram
        ax = axes[i]
        ax.hist(data, bins=bins, alpha=0.7, color=color)

        # Create titles and labels
        if titles is not None:
            ax.set_title(titles[i])
        ax.set_xlabel('Score')
        ax.set_ylabel('Frequency')

    plt.tight_layout()
    plt.show()
```

```python
# Define score columns
rd_1_columns = ['var 1.1', 'var 1.2', 'tot 1']
rd_2_columns = ['var 2.1 ', 'var 2.2', 'tot 2'] 
rd_3_columns = ['var 3.1', 'var 3.2', 'var 3.3', 'tot 3']
rd_4_columns = [ 'var 4.1', 'var 4.2', 'var 4.3', 'var 4.4', 'tot 4']
rd_5_columns = ['var 5.1 ', 'var 5.2', 'var 5.3', 'tot 5']
rd_6_columns = ['var 6.1', 'var 6.2', 'var 6.3', 'var 6.4', 'tot 6']
total_columns = [rd_1_columns, rd_2_columns, rd_3_columns, rd_4_columns, rd_5_columns, rd_6_columns]

# Create histograms for each risk driver 
for n, cols in enumerate(total_columns):
    print('Risk driver ' + str(n + 1))
    data_lists = [df_total[c] for c in cols]
    titles = cols
    plot_histograms_in_row(data_lists, titles=titles, bins=20)
```
